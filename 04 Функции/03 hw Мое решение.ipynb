```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []}

def get_dir(doc_nr = ''):
    if doc_nr == '':
        doc_nr = input('Введите номер документа: ')
    for di in directories.items():
        if doc_nr in di[1]:
            return(di[0])
    return('Документ не найден в базе')

def get_doc():
    doc_nr = input('Введите номер документа: ')
    for do in documents:
        if doc_nr in do['number']:
            return(do['name'])
    return('Документ не найден в базе')

def get_all():
    res = ""
    for do in documents:
        print('№: '+ do['number'] + ', тип: ' + do['type'] + ', владелец: ' + \
              do['name'] + ', полка хранения: ' + get_dir(do['number']))

def del_dir():
    dir_nr = input('Введите полку: ')
    for di in directories.items():
        if dir_nr in di[0]:
            if di[1] == []:
                del(directories[dir_nr])
                print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys()) + '.')
                return
            else:
                print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' \
                       + ', '.join(directories.keys()) + '.')
                return
    print('Полки ' + dir_nr + ' нет.')

def add_doc():
    nr = input('Введите номер документа: ')
    typ = input('Введите тип документа: ')
    ppl = input('Введите владельца документа: ')
    di = input('Введите полку для хранения: ')
    documents.append({'type': typ, 'number': nr, 'name': ppl})
    if di in directories.keys():
        print(di)
        directories[di].append(nr)
    else:
        directories[di] = [nr]
    print('Документ добавлен. Текущий список документов: ')
    get_all()

def del_doc():
    doc_nr = input('Введите номер документа: ')
    str = ''
    for do in documents:
        if doc_nr in do['number']:
            documents.remove(do)
            str = 'Документ удален.'
        else:
            str = 'Документ не найден в базе.'
    print(str + ' Текущий список документов:')
    get_all()

def move_doc():
    nr = input('Введите номер документа: ')
    di = input('Введите номер полки: ')
    di_old =  get_dir(nr)
    if di_old == 'Документ не найден в базе':
        print('Документ не найден в базе. Текущий список документов:')
        get_all()
    elif di not in directories:
        print('Такой полки не существует. Текущий перечень полок: ' + ', '.join(directories.keys()))
    else:
        directories[di_old].remove(nr)
        directories[di].append(nr)
        print('Документ перемещен. Текущий список документов:')
        get_all()

inp_str = ''
while inp_str != 'q':
    inp_str = input('Введите команду: ')
    if inp_str == 's':
        print('Документ хранится на полке: ' + get_dir())
    elif inp_str == 'p':
        print('Владелец документа: ' + get_doc())
    elif inp_str == 'l':
        get_all()
    elif inp_str == 'ds':
        del_dir()
    elif inp_str == 'ad':
        add_doc()
    elif inp_str == 'd':
        del_doc()
    elif inp_str == 'm':
        move_doc()
```
