## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re

def test_num():
    num = input('Номер: ')
    pat = r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}'
    res = re.match(pat, num)
    if res is not None:
        print(f'Номер {res.group()[:6]} валиден. Регион: {res.group()[6:8]}')
    else:
        print(f'Номер не валиден.')
test_num()
````

## Задание 2

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат:

````python
some_string = 'Напишите функцию функцию, которая которая будет будет будет будет удалять удалять удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'

def fix_str(str):
    return re.sub(r'(\w+)\s\1+', r'\1',re.sub(r'(\w+)\s\1+', r'\1', str))
print(fix_str('1 2 3 3 4 5 5 6'))
print(fix_str(some_string))
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
def akro(s):
    st = ''.join(re.findall(r'^\w', s) + re.findall(r'(?<=\s)\w', s))
    print( st.upper())
akro('Информационные технологии')
akro('Near Field Communication')
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
def rasp(mas):
    dom = re.findall(r'(?<=@)\w+\.\w{2,3}', ' '.join(mas))
    dom = list(set(dom))
    coun = []
    for i in range(0, len(dom)):
        coun.append(0)
    for m in mas:
        for i, d in enumerate(dom):
            if re.search(d, m) is not None:
                coun[i] +=  1
    for i, d in enumerate(dom):
        print(f'{dom[i]}: {coun[i]}')
rasp(emails)
````

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Результат: 
```

Слов на согласные буквы: 21 
```

````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
def count_a(s):
    a = re.findall('^[АЯУЕИЮЭЫЁОоаяуюиэёеиыEYUIOAeyuioa]', s) + re.findall('\s[АЯУЕИЮЭЫЁОоаяуюиэёеиыEYUIOAeyuioa]', s)
    b = re.findall('^[^АЯУЕИЮЭЫЁОоаяуюиэёеиыEYUIOAeyuioa]', s) + re.findall('\s[^АЯУЕИЮЭЫЁОоаяуюиэёеиыEYUIOAeyuioa]', s)
    print(f'Слов на гласные буквы: {len(a)}')
    print(f'Слов на согласные буквы: {len(b)}')
count_a(some_text)
````
